```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/laptop_price.csv",encoding = "latin-1")
df
```

<h3>Types of Columns</h3>

```python
numerical_features = [col for col in df.columns if df[col].dtype != "object"]
categorical_features = [col for col in df.columns if df[col].dtype == "object"]
print("numerical_features :",numerical_features)
print("categorical_features :",categorical_features)
```

<h3> Data Analysis</h3>

```python
fig,axs = plt.subplots(1,2,figsize = (15,7))
plt.subplot(121)
sns.histplot(data = df,x = "Price_euros",kde = True,bins = 30,color = "green")
plt.title("Price in Euro")
plt.subplot(122)
sns.histplot(data = df,x = "Price_euros",kde = True,hue = "Ram")
plt.title("Price in Euro and Ram")
plt.show()
```

<h3>Observations<h3>
<h4>Laptops with 4GB and 8GB Rams are in most trending</h4>

```python
fig,axs = plt.subplots(1,2,figsize = (15,7))
plt.subplot(121)
sns.histplot(data = df,x = "Price_euros",kde = True,bins = 30,color = "green")
plt.title("Price in Euro")
plt.subplot(122)
sns.histplot(data = df,x = "Price_euros",kde = True,hue = "OpSys")
plt.title("Price in Euro and Operating system")
plt.show()
```

<h4>Observation</h4>
<h5> Windows operating system has huge demand</h5>

```python
fig,axs = plt.subplots(1,2,figsize = (15,7))
plt.subplot(121)
sns.histplot(data = df,x = "Price_euros",kde = True,bins = 30,color = "green")
plt.title("Price in Euro")
plt.subplot(122)
sns.histplot(data = df,x = "Price_euros",kde = True,hue = "TypeName")
plt.title("Price in Euro and TypeName")
plt.show()
```

<h4>Observations</h4>
<h5> Notebooks are at low price and Gaming laptops are at higher price </h5>

```python
sns.scatterplot(data = df,x = "Ram",y = "Price_euros",hue = "TypeName")
plt.title("Ram and Price in Euros")
plt.show()
```

<h4>Observations</h4>
<h5>As ram size increases Price increases</h5>

```python
plt.rcParams['figure.figsize'] = (30,12)
plt.subplot(1,5,1)
size = df["Ram"].value_counts()
labels = "8GB","16GB","4GB","2GB","12GB","6GB","32GB","24GB","64GB" 
color_names = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'pink', 'cyan', 'magenta']
plt.pie(size,colors = color_names,labels = labels,autopct = "%2f%%")
plt.title("Ram",fontsize = 20)
plt.axis("off")
```

<h3>Observations</h3>
<h5> Laptops with 4GB,8GB,16GB Rams are in  demand</h5>

```python
plt.rcParams['figure.figsize'] = (30,12)
plt.subplot(1,5,1)
size = df["OpSys"].value_counts()
labels =  'Windows 10', 'No OS', 'Linux', 'Windows 7', 'Chrome OS', 'macOS','Mac OS X', 'Windows 10 S', 'Android'
color_names = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'pink', 'cyan', 'magenta']
plt.pie(size,colors = color_names,labels = labels,autopct = "%2f%%")
plt.title("OpSys",fontsize = 20)
plt.axis("off")
```

<h4>Observations</h4>
<h5> Laptops with Windows10 operating system are in demand</h5>

```python
plt.rcParams['figure.figsize'] = (30,12)
plt.subplot(1,5,1)
size = df["TypeName"].value_counts()
labels =   'Notebook', 'Gaming', 'Ultrabook', '2 in 1 Convertible', 'Workstation','Netbook'
color_names = ['red', 'blue', 'green', 'yellow', 'orange', 'purple' ]
plt.pie(size,colors = color_names,labels = labels,autopct = "%2f%%")
plt.title("TypeName",fontsize = 20)
plt.axis("off")
```

<h4>Observations</h4>
